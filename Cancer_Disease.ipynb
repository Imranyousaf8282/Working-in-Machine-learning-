### **Feature Enginnering**

```python
cancer_data['Progesterone Status']=='Positive'
```

```python
cancer_data['Progesterone Status'] = cancer_data['Progesterone Status']=='Positive'
```

```python
cancer_data['A Stage']= cancer_data['A Stage'] =='Regional'
```

```python
cancer_data['Estrogen Status']= cancer_data['Estrogen Status']=='Positive'
```

```python
cancer_data['Status'] = cancer_data['Status'] =='Alive'
```

```python
cancer_data.info()
```

```python
cancer_data.select_dtypes(["O"]).columns
```

```python
cancer_data['Race'].value_counts()
```

```python
cancer_data['Race'].unique()
```

```python
cancer_data['Race'] = cancer_data['Race'].replace(['White', 'Black', 'Other'],[1,0,2])
```

```python
cancer_data['Marital Status'].unique()
```

```python
cancer_data['Marital Status']=cancer_data['Marital Status'].replace(['Married', 'Divorced', 'Single ', 'Widowed', 'Separated'],[1,0,2,3,4])
```

```python
cancer_data['T Stage '].unique()
```

```python
cancer_data['T Stage ']=cancer_data['T Stage '].replace(['T1', 'T2', 'T3', 'T4'],[1,2,3,4])
```

```python
cancer_data['N Stage'].unique()
```

```python
cancer_data['N Stage']=cancer_data['N Stage'].replace(['N1', 'N2', 'N3'],[1,2,3])
```

```python
cancer_data['6th Stage'].unique()
```

```python
cancer_data['6th Stage']=cancer_data['6th Stage'].replace(['IIA', 'IIIA', 'IIIC', 'IIB', 'IIIB'],[1,2,3,4,5])
```

```python
cancer_data['Grade'].unique()
```

```python
cancer_data['Grade']=cancer_data['Grade'].replace(['3', '2', '1', ' anaplastic; Grade IV'],[1,2,3,4])
```

```python
cancer_data['differentiate'].unique()
```

```python
cancer_data['differentiate']=cancer_data['differentiate'].replace(['Poorly differentiated', 'Moderately differentiated',
       'Well differentiated', 'Undifferentiated'],[1,2,3,0])
```

```python
cancer_data.head()
```

```python
cancer_data.info()
```

```python
cancer_data.corr()
```

```python
X = cancer_data.drop("Status",axis=1)
y = cancer_data["Status"]
```

```python
X.info()
```

### Training of model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=101)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
L_regression = LogisticRegression()
```

```python
L_regression.fit(X_train,y_train)
```

```python
prediction_train = L_regression.predict(X_train)
```

```python
prediction_test = L_regression.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score
```

```python
accuracy_score(y_train,prediction_train)
```

```python
accu_score = accuracy_score(y_test,prediction_test)
accu_score*100
```

```python
print(classification_report(y_test,prediction_test))
```

```python
preci_score = precision_score(y_test,prediction_test)
preci_score
```

```python
recal_score = recall_score(y_test,prediction_test)
recal_score
```

```python
cm = confusion_matrix(y_test,prediction_test)
sns.heatmap(cm,annot=True,fmt='d')
```
