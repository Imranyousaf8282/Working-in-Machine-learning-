```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic = pd.read_csv('kashti.csv')
titanic
```

```python
titanic.info()
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
titanic.rename({"Unnamed: 0":"passengerId"},axis=1,inplace=True)
```

```python
titanic
```

```python
titanic.drop("alone",axis=1,inplace=True)
titanic.drop("class",axis=1,inplace=True)
titanic.drop("who",axis=1,inplace=True)
titanic.drop("adult_male",axis=1,inplace=True)
titanic.drop("deck",axis=1,inplace=True)
titanic.drop("embark_town",axis=1,inplace=True)
titanic.drop("alive",axis=1,inplace=True)
```

```python
titanic
```

```python
titanic.shape
```

```python
titanic.select_dtypes(["O"]).columns
```

```python
sex = pd.get_dummies(titanic['sex'],drop_first=True)
embarked = pd.get_dummies(titanic['embarked'], drop_first=True)
```

```python
titanic.head()
```

```python
titanic = pd.concat([titanic,sex,embarked],axis=1)
```

```python
titanic.head()
```

```python
titanic.drop(['sex','embarked'],axis=1,inplace=True)
```

```python
titanic
```

```python
titanic.select_dtypes(["O"]).columns
```

```python
titanic['age'].fillna(titanic['age'].mode()[0], inplace=True)
```

```python
titanic
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
predictors = ["passengerId","pclass","age","sibsp","parch","fare","male","Q","S"]
```

```python
X = titanic[predictors]
y = titanic['survived']
```

### model training

```python
from sklearn.model_selection import train_test_split
X_trian,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=101)
```

## Logistic regresssion 

```python
from sklearn.linear_model import LogisticRegression
lreg = LogisticRegression()
```

```python
lreg.fit(X_trian,y_train)
```

```python
lreg
```

```python
prediction_train = lreg.predict(X_trian)
```

```python
prediction_test = lreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,recall_score,precision_score
```

```python
acc_score_train = accuracy_score(y_train,prediction_train)
acc_score_train*100
```

```python
acc_score_test = accuracy_score(y_test,prediction_test)
acc_score_test*100
```

```python
class_report_train = classification_report(y_train,prediction_train)
print(class_report_train)
```

```python
class_report_test = classification_report(y_test,prediction_test)
print(class_report_test)
```

```python
cm_train = confusion_matrix(y_train,prediction_train)
sns.heatmap(cm_train,annot=True,fmt="d")
```

```python
cm_test = confusion_matrix(y_test,prediction_test)
sns.heatmap(cm_test,annot=True,fmt="d")
```

```python
recall_score_train = recall_score(y_train,prediction_train)
recall_score_train
```

```python
recall_score_test = recall_score(y_test,prediction_test)
recall_score_test
```

```python
precision_score(y_train,prediction_train)
```

```python
precision_score(y_test,prediction_test)
```

* Precision_score = Tp(Tp+Fp)
* Recall_score = Tp(Tp+FN)
* F1_score = (precision_ppv*value + precision_ppn*value)/total_value
