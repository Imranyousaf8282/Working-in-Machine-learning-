```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
usa_housing = pd.read_csv("USA_Housing.csv")
usa_housing
```

### **TRAINING DATA PRE-PROCESSING**
The first step in the machine learning pipeline is to clean and transform the training data into a useable format for analysis and modeling.   

As such, data pre-processing addresses:
- Assumptions about data shape
- Incorrect data types
- Outliers or errors
- Missing values
- Categorical variables

```python
usa_housing.shape
```

```python
usa_housing.info()
```

```python
sns.heatmap(usa_housing.isnull())
plt.title("Missing Data : Training Set")
plt.show()
```

```python
# drop address colum in data
usa_housing.drop("Address",axis=1,inplace=True)
```

```python
usa_housing
```

```python
usa_housing.isnull().sum()
```

```python
usa_housing.dropna(inplace=True)
```

```python
usa_housing
```

### Data is ready for applying model


```python
usa_housing.shape
```

## **Target Variable Splitting**
We will spilt the Full dataset into **Input** and **target** variables

Input is also called **Feature Variables**
Output referes to Target **variables**

```python
predictors =["Avg. Area Income","Avg. Area House Age","Avg. Area Number of Rooms","Avg. Area Number of Bedrooms","Area Population"]
predictors
```

```python
X = usa_housing[predictors]
y = usa_housing["Price"]
```

#### Model Training

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state=101)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

```python
linear_reg.fit(X_train,y_train)
```

```python
y_pred_train = linear_reg.predict(X_train)
y_pred_test = linear_reg.predict(X_test)
```

```python
usa_housing.hist("Price")
```

```python
sns.scatterplot(x=y_test,y=y_pred_test,color="blue",label="Actual Data Points")
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color="red",label='ideal line')
plt.legend()
plt.show()
```

```python
# combine actual and pred value
result = np.column_stack((y_test,y_pred_test))
print("Actual Value | Predicted value")
print("..................................")

for actual,predicted in result:
    print(f"{actual:14.2f} | {predicted:12.2f}")
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y_train,y_pred_train)
mse
```

```python
mse = mean_squared_error(y_test,y_pred_test)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2_sc=r2_score(y_test,y_pred_test)
r2_sc
```

```python
residual = actual - y_pred_test.reshape(-1)
print(residual)
```

```python
residual = y_train - y_pred_train
residual.sum().round()
```

```python
plt.scatter(y_pred_train,residual)
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train,y_train)
```

```python
y_pred_dtr = dt_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred_dtr)
mse
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train,y_train)
y_pred_rfr = rf_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_rfr = mean_squared_error(y_test,y_pred_rfr)
mse_rfr
```

## Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_regr = GradientBoostingRegressor()
gb_regr.fit(X_train,y_train)
```

```python
y_pred_gbr = gb_regr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_gbr = mean_squared_error(y_test,y_pred_gbr)
mse_gbr
```
